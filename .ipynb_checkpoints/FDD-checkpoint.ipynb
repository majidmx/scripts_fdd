```python
import smtplib
from email.mime.text import MIMEText
from email.mime.multipart import MIMEMultipart

def send_email(from_address, to_address, subject, email_body):

    msg = MIMEMultipart()
    msg['From'] = from_address
    msg['To'] = to_address
    msg['Subject'] = subject

    msg.attach(MIMEText(email_body, 'plain'))

    text = msg.as_string()

    try:
        smtpObj = smtplib.SMTP('smtp.amazon.com')
        smtpObj.sendmail(from_address, to_address, text)         
        print("Successfully sent email.")

    except smtplib.SMTPException as smtpexception:
        print("Error: unable to send email due to: %s" % smtpexception)

from_address = 'majidmm@amazon.com'
to_address = 'majidmm@amazon.com'
subject = 'test email'
email_body = 'This is a test'

send_email(from_address, to_address, subject, email_body)
```
